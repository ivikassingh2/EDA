#### Categorial Variables Analysis

# Insight #1 This code creates a bar chart visualizing the number of movies for each genre in the bollywood_df DataFrame

```python
plt.figure(figsize=(10,6))
ax = sns.countplot(data=bollywood_df, x='Genre', order=bollywood_df['Genre'].value_counts().index, palette='Set2')
plt.title('Number of Movies per Genre')
plt.xlabel('Genre')
plt.ylabel('Count')
# Adding the data labels the bar numbers
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')
plt.xticks(rotation=45)
plt.legend(title='Genre',labels=bollywood_df['Genre'])
plt.show()

#This code creates a bar chart visualizing the number of movies for each genre in the bollywood_df DataFrame
```

# Insight #2 This code creates a bar chart that visualizes the number of movies released in each month based on the bollywood_df DataFrame.

```python
plt.figure(figsize=(10,8))
sns.countplot(data=bollywood_df, x='Month', order=sorted(bollywood_df['Month'].unique()), palette='Set3')
plt.title('Number of Movies Released per Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.legend(bollywood_df['Month'].unique())
plt.show()

#This code creates a bar chart that visualizes the number of movies released in each month based on the bollywood_df DataFrame.
```

# Insight #3 Detailed bar chart to visualize the number of movies released each month in all the years with datalabels

```python
plt.figure(figsize=(10, 8))

# Define month names and a better color palette
month_names = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
palette = sns.color_palette('pastel')

# Create the count plot
ax = sns.countplot(data=bollywood_df, x='Month', order=sorted(bollywood_df['Month'].unique()), palette=palette)
ax.set_xticklabels(month_names, rotation=45)

# Adding the data labels the bar numbers
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')

plt.title('Number of Movies Released per Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.legend(title='Months', labels=month_names, loc='upper right')
plt.show()

#This code creates a detailed bar chart to visualize the number of movies released each month in all the years with datalabels
```

# Insight #4 This code generates a bar chart to visualize the number of movies released each year.

```python
plt.figure(figsize=(10, 6))

# Create the count plot and assign it to ax
ax = sns.countplot(data=bollywood_df, x='Year', order=sorted(bollywood_df['Year'].unique()), palette='Set1')

# Adding counts above the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')

plt.title('Number of Movies Released per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Year', labels=sorted(bollywood_df['Year'].unique()), loc='upper right')  # Updated legend
plt.show()

#This code generates a bar chart to visualize the number of movies released each year.
```

# Insight #21 Top five genres by total box office collection in Bollywood films.

# Insight #23 Top 10 Bollywood movies by YouTube views.
# The horizontal bar plot effectively communicates which movies have garnered the most views on YouTube, with annotations formatted in millions for clarity. The x-axis is also formatted to enhance readability by showing the view counts in millions.

```python
from matplotlib.ticker import FuncFormatter

top_movies = bollywood_df.sort_values(by='YoutubeViews', ascending=False).head(10)
plt.figure(figsize=(12,8))

# Plotting the bar chart
ax = sns.barplot(x='YoutubeViews', y='MovieName', data=top_movies, palette='viridis')

# Adding data labels with formatting (in millions)
for index, value in enumerate(top_movies['YoutubeViews']):
    plt.text(value, index, f'{value/1e6:.2f}M', va='center', ha='left', color='black')

# Formatting the x-axis to show values in millions
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))

plt.title('Top 10 Movies by YouTube Views')
plt.xlabel('YouTube Views (in Millions)')
plt.ylabel('Movie Name')
plt.show()

# This code calculates and visualizes the top 10 Bollywood movies by YouTube views.
# The horizontal bar plot effectively communicates which movies have garnered the most views on YouTube, with annotations formatted in millions for clarity. The x-axis is also formatted to enhance readability by showing the view counts in millions.
```

# Insight #28 YouTube "Like Ratio" for Bollywood movies, addressing potential division by zero issues, and then visualizes the distribution of this ratio using a histogram.
# The plot helps to understand how audience engagement (likes relative to views) varies across different movies, which can provide insights into their popularity and reception.

```python
bollywood_df['LikeRatio'] = bollywood_df['YoutubeLikes'] / bollywood_df['YoutubeViews']

# Handle division by zero if any
bollywood_df.replace([np.inf, -np.inf], 0, inplace=True)

# Plot Like Ratio
plt.figure(figsize=(10,6))
sns.histplot(bollywood_df['LikeRatio'], bins=20, kde=True, color='blue')
plt.title('YouTube Like Ratio Distribution')
plt.xlabel('Like Ratio')
plt.ylabel('Frequency')
plt.show()

# This code calculates the YouTube "Like Ratio" for Bollywood movies, addressing potential division by zero issues, and then visualizes the distribution of this ratio using a histogram.
# The plot helps to understand how audience engagement (likes relative to views) varies across different movies, which can provide insights into their popularity and reception.
```

# Insight #29 YouTube "Dislike Ratio" for Bollywood movies, handles potential division by zero issues, and visualizes the distribution of this ratio using a histogram.
# The plot provides insights into how audience dissatisfaction (dislikes relative to views) varies across different movies, allowing for an analysis of viewer sentiment and engagement.

```python
bollywood_df['DislikeRatio'] = bollywood_df['YoutubeDislikes'] / bollywood_df['YoutubeViews']

# Handle division by zero if any
bollywood_df.replace([np.inf, -np.inf], 0, inplace=True)
# Plot Dislike Ratio
plt.figure(figsize=(10,6))
sns.histplot(bollywood_df['DislikeRatio'], bins=20, kde=True, color='red')
plt.title('YouTube Dislike Ratio Distribution')
plt.xlabel('Dislike Ratio')
plt.ylabel('Frequency')
plt.show()

# This code calculates the YouTube "Dislike Ratio" for Bollywood movies, handles potential division by zero issues, and visualizes the distribution of this ratio using a histogram.
# The plot provides insights into how audience dissatisfaction (dislikes relative to views) varies across different movies, allowing for an analysis of viewer sentiment and engagement.
```
